5.python program to upload image and perform different task on it

```python
from google.colab import files
from PIL import Image
import io
import matplotlib.pyplot as plt

uploaded = files.upload()

filename = next(iter(uploaded))

image = Image.open(io.BytesIO(uploaded[filename]))

plt.figure(figsize=(6,6))
plt.title('Original Image')
plt.imshow(image)
plt.axis('off')
plt.show()

gray_image = image.convert('L')

plt.figure(figsize=(6,6))
plt.title('Grayscale Image')
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.show()

resized_image = image.resize((100, 100))

plt.figure(figsize=(4,4))
plt.title('Resized Image (100x100)')
plt.imshow(resized_image)
plt.axis('off')
plt.show()
```

6.python prgm to take video and break it into frames


```python

!pip install opencv-python-headless


from google.colab import files
uploaded = files.upload()


import cv2
import os
import shutil


video_path = next(iter(uploaded))
output_folder = "frames"
os.makedirs(output_folder, exist_ok=True)


cap = cv2.VideoCapture(video_path)
frame_number = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break
    frame_filename = f"{output_folder}/frame_{frame_number:04d}.jpg"
    cv2.imwrite(frame_filename, frame)
    frame_number += 1

cap.release()
print(f" Total frames extracted: {frame_number}")

shutil.make_archive("frames_download", 'zip', output_folder)


files.download("frames_download.zip")
```

7.python pgrm to read and display content/no. of the file

```python

file_name = "example.txt"
try:
    with open(file_name, "r") as file:
        content = file.read()
        print("File Content:\n")
        print(content)

    with open(file_name, "r") as file:
        lines = file.readlines()
        print("\nNumber of lines in file:", len(lines))

except FileNotFoundError:
    print("Error: File not found.")
```
